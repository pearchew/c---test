### Project Title: The High-Speed Options Pricer (Black-Scholes Model)
**The Goal:** Write a C++ program from scratch that calculates the fair price of a "European Call Option" (a financial contract) using the famous Black-Scholes mathematical formula. 

While the math sounds advanced, programmatically, it is just a calculator. This is the perfect beginner project because it teaches you how C++ handles basic math, data types, and functions without getting bogged down in complex memory management.

Here is your step-by-step roadmap:

### Phase 1: The "Boilerplate" & Strict Data Types
Unlike Python, where you just open a file and start typing `price = 100`, C++ requires a structural shell and strict declarations of what kind of data you are holding.
*   **The Task:** Set up your C++ file, include the basic input/output libraries, and declare your financial variables.
*   **Key C++ Concepts to Learn:**
    *   `#include <iostream>`: This tells C++ you want to use the standard library for printing text to the screen and taking user input.
    *   `int main() { ... }`: **Every** C++ program must have a "main" function. This is where the execution starts.
    *   **Strict Typing (`double` vs `int`):** In C++, you cannot just write `StockPrice = 100.50`. You must tell the computer exactly how much memory to allocate. You will learn to use `double StockPrice = 100.50;` (for decimals) and `int TimeToMaturity = 1;` (for whole numbers).

### Phase 2: Building a Math Function (`<cmath>`)
In Python, you might import `scipy` to get complex statistical functions. In C++, you build them yourself using the standard math library. The Black-Scholes formula requires calculating a "Cumulative Distribution Function" (CDF).
*   **The Task:** Write a standalone C++ function that calculates the Normal CDF using the math library.
*   **Key C++ Concepts to Learn:**
    *   `#include <cmath>`: The library that gives you access to square roots (`sqrt`), logarithms (`log`), and exponentials (`exp`).
    *   **Function Signatures:** In Python, you write `def calculate_cdf(x):`. In C++, you must declare what type of data the function will return. You will write `double calculate_cdf(double x) { ... }`.
    *   **The `std::erf` function:** You will use the C++ "error function" to construct your statistical distribution.

### Phase 3: Implementing the Black-Scholes Logic
Now you plug your financial variables into the formula. This phase teaches you how C++ executes sequential logic and arithmetic.
*   **The Task:** Write the formula to calculate "d1" and "d2" (the core components of Black-Scholes), and then calculate the final Call Option price.
*   **Key C++ Concepts to Learn:**
    *   **Scope:** Understanding that a variable created inside your `calculate_cdf` function cannot be seen or used by your `main` function. 
    *   **Arithmetic Precision:** Learning how C++ handles division. (e.g., in C++, `1 / 2` equals `0`, not `0.5`, because both are integers. You will learn to write `1.0 / 2.0` to force decimal precision—a common beginner trap!).

### Phase 4: User Interaction & Compilation
You want a user (or a trader) to be able to type in a stock price and instantly get the option price back. 
*   **The Task:** Make the program interactive and convert your code into an executable application.
*   **Key C++ Concepts to Learn:**
    *   `std::cout`: (Character Output) How to print questions to the screen. E.g., `std::cout << "Enter the current stock price: ";`
    *   `std::cin`: (Character Input) How to pause the program and wait for the user to type a number, saving it to a variable.
    *   **Compiling (The biggest difference from Python):** Python reads your code line-by-line. C++ must be "compiled" (translated entirely into machine code/binary) before it runs. You will learn how to use a compiler (like `g++` or `clang++` in your terminal) to turn your `.cpp` text file into a `.exe` or executable app.

### Why this is the perfect gateway to your Master's:
1.  **Direct Curriculum Overlap:** When you take the traditional finance modules, you will be taught the Black-Scholes model. Building it in code demystifies the math. When you see the formula on a whiteboard, you’ll mentally map it to your C++ functions.
2.  **Understanding "The Stack":** By doing this, you will understand the FinTech ecosystem. You will realize that you could use your Python/Pandas project (from the previous prompt) to *find* the historical volatility of a stock, and then feed that volatility into this C++ program to *price* the derivative instantly.
3.  **A Low-Stakes Intro:** C++ is notorious for being difficult because of "pointers" and "memory leaks". This specific project avoids memory management entirely. It lets you get comfortable with the strict syntax, compiling, and typing before Imperial throws the truly hard computer science concepts at you in the electives.
